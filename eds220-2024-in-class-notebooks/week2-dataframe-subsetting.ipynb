## Selecting a single column

Simplest case: select a single column by column name

General syntax:
```python
df['column_name']
```

This is an example of **label based subsetting** which means we want to select data from out ddf using the *names* of the cols, not their position

### Example
Select number of bird species observed at Mugu Lagoon in spring

````python
# select a single col by using square brackets []
mul_spring = df['MUL_spring']

mul_spring
````

Recall: we can think of a `pandas.DataFrame` as a dict of its cols. 
Then, we can access a single column using the column name as the *key*, just like we would in a dictionary

````python
df.MUL_spring
````

**Favor `df['column_name']` instead of `df.column_name`**

why?

- `df['column_name']` avoids conflights with `pd.DataFrames` methods and attributes. For ex, if `df` has a col named `count` it's ambiguous whether `pd.count` is referring to the `count()` method or the `count` column
- `df['column_name']` can take any col name
- `df.col_name` only works if the column has no spaces or special characters

## Selecting multiple columns... 

### using a list of col names

Select mult cols in a single call by passing a list of col names to square brackets `[]`

```python
df[['col1', 'col2', 'col100']]
```

Notice the "double square brackets". This is becuase we are passing the list of names `['col1', 'col2', 'col100']` to the selection brackets `[]`

## Check in

Is this label based or location based? This is label based

````python
# Select columns with names "TJE_winter" and "TJE_fall"
tje_wf = df[['TJE_winter','TJE_fall']]
````

### using a slice

To select a slice of the columns we will use a special case of ** `loc` selection**, General syntax:
```python
df.loc[ :, 'column_start':'column_end']
```
`column_start` and `column_end` = the starting and ending points of the col slice we want to subset from the data frame.

notice:
- the first value passed to `loc` is used for selecting rows, using a colon `:` as the row-selection parameter means "select all the rows"
- the slice of the resulting df includes both endpoints of the slice

````python
# Select columns between 'CSM_winter' and 'MUL_fall'

csm_mul = df.loc[:, 'CSM_winter': 'MUL_fall']
csm_mul
````

## Selecting rows

### using a condition
to sel rows that satify a particular condition the general syntax is 
```python
df[condition_on_rows]
```

The `condition_on_rows` can be many things, let's see some usual scenarios.

### Example:
We are interested in data after 2020

````python
# select all rows with year >2020
post_2020 = df[ df['year'] > 2020 ]
post_2020
````

condition for our rows = `df['year'] > 2020`
this is a `pandas.Series` with booooolean values (`True` or `False`) indicating which rows satisfy the condition year > 2020

````python
# check the type of df['year'] > 2020
print(type(df['year'] > 2020))

# print the boolean series
df['year'] > 2020
````

This series is acting as a mask for the df

When we pass such a series of boolean vals to the sel brackets, we only keep the rows that correspond to a `True` value.

## Check in

Get the subset of the df on which the San Dieguito Wetland has at least 75 species recorded during spring

````python
sdw_spring = df[df['SDW_spring'] >= 75]

sdw_spring
````

### Example

We want to look at data from years 2012 to 2015 

````python
subset = df[df['year'].between(2012,2015)]
subset
````

Let's break it down:

1. df['year'] = column with the year values, a `pandas.Series`
2. in `df['year'].between()` we are acessiong a method for the `pandas.Series` using `,`
3. (2012, 2015) = parameters, this includes both endpoints
4. `df['year'].between(2012, 2015)` = `pandas.Series` of boolean values indicating which rows have year value between 2012 and 2015
5. When we put `df['year'].between(2012, 2015)` inside the selection brackets `[]` we obtain the rows we wanted to select

### Avoid using `loc` for selecting only rows

It is equivaluent to write the following two:

````python
# Select rows with year<2015
df[df['year']<2015]
````

````python
# Select rows with year<2015 using loc
df.loc[ df['year']<2015 , :]
````

### using multiple conditions

We can combine multiple conditions to select rows by surrounding each condition in parentesis `()` and using the or operator `|` and the and operator `&`.

#### Example: or
Select rows in whichCSM has more than 50 species registered in winter *or* fall

````python
df[ (df['CSM_winter']>60) | (df['CSM_fall']>60) ]
````

An empty data frame! That's ok, it just means no rows satisfy the conditions.

## Select rows by position

Sometimes we may want to sel certain rows based on their *actual position* in te data frame. In other workds, using **position-based subsetting**

To do this, we use **iloc selection** syntax:

```python
df.iloc[row-indicies]
```

`iloc` = stands for integer-location based indexing

#### Example:

````python
# Select the fifth row (index=4)
df.iloc[4]
````

````python
# Select rows from the 10th through the 13th, including 13
df.iloc[9:14]
````

### Selecting rows and columns simultaneously

Can be done using `loc` (labels) or `iloc` (positions)

### ... by labels or conditions
When we want to select rows and columns simultaneously by labels (including using conditions) we can use `loc` selection with the syntax:

```python
df.loc[ row-selection, column-selection]
```

Example: 

select winter surveys from MUL and TJE after 2020

````python
df.loc[df['year'] > 2020, ['MUL_winter', 'TJE_winter']]
````

### ... by position

Suppose we want to select rows 3-7 (including 7) and 4th and 5th column

````python
df.iloc[3:8, [3,4]]
````

## Notes about `iloc` and `loc`

### Which one does what?

Remember that the `i` in `iloc` stands for integer location and it uses integer indexing the same way as indexing for Python list.

### `iloc` for columns selection? Avoid it!

We can also access columns be position using `iloc`, but it's best to avoid it

Example:

We want to access the 9th column of the df and we want to select a column by its position.

````python
df.iloc[:,8]
````

Unless you are *really* looking for info about the 9th column, **do not acess a column by position**. This can break in many ways:

1. it relies on you correctly counting the position of a column. Even with a small dataset this can be prone to error

2. It is not explicit. If we want info about that column, then we use `[df['SDW_spring']`
